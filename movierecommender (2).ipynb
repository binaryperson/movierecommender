```python
import numpy as np
import pandas as pd
import ast
import sklearn
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=movies.merge(credits,on='title')
```

```python
#preprocessing begins fro here
#genres
#id
#keywords
#title
#overview
#cast
#crew
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter +=1
        else:
            break   
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
        return L
            
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x] if x is not None else [])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x] if x is not None else [])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x] if x is not None else [])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x] if x is not None else [])
```

```python
movies.head()
```

```python
# Convert lists to strings using join() before concatenation
movies['tags'] = movies['overview'] + \
                 movies['crew'].apply(lambda x: ', '.join(x)) + \
                 movies['keywords'].apply(lambda x: ', '.join(x)) + \
                 movies['cast'].apply(lambda x: ', '.join(x)) + \
                 movies['genres'].apply(lambda x: ', '.join(x))
```

```python
def stem(text):
    y=[]
    for i in text.split():#convertingstringtolist
        y.append(ps.stem(i))
    return " ".join(y) #final string
```

```python
#calculating cosine(angle) distance not euclidean distance
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
list(enumerate(similarity[0])) #output as seen list converted into list of tuples
#done so that we don't lose the index position when the lsit is sorted
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0] #masked
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        
        print(new_df.iloc[i[0]].title)
    return
```

```python
recommend('')
```

```python
recommend('Spider-Man 3')
```
